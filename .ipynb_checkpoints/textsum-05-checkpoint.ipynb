## 文本摘要综述之论文笔记

> [Abstractive Text Summarization using Sequence-to-sequence RNNs and Beyond](https://arxiv.org/pdf/1602.06023.pdf)(2016年)

### 模型

#### 1.1 Encoder-Decoder RNN with Attention and Large Vocabulary Trick

模型是在神经机器翻译模型(Bahdanau et.al)的基础上做的改进.

- 编码层: 双向GRU-RNN(Chung et.al)
- 解码层: 单向GRU-RNN

引入 Large Vocablary Trick 来解决 decoder 词表过大的问题.

> [tensorflow中的实现](https://www.tensorflow.org/api_docs/python/tf/nn/sampled_softmax_loss)

> [参考论文](https://arxiv.org/abs/1412.2007)

#### 1.2 capturing Keywords using Feature-rich Encoder

加入传统的 TF-IDF，POS，NER 等特征来尝试抓住句子的关键部分



#### 1.3 Modeling Rare/Unseen Words using Switching Generator-Pointer

引入 Generator-Pointer 来解决 OOV 和低频词的问题



#### 1.4 Capturing Hierarchical Document Structure with Hierarchical Attention

引入 Hierarchical Attention 来抓住句子的重要性信息

> [参考论文](https://arxiv.org/abs/1506.01057)



#### 1.5 引入 Temporal Attention 来缓解连续生成重复词的问题

> [参考论文](https://arxiv.org/abs/1608.02927)
